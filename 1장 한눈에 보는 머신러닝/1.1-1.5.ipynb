### 1.1 머신러닝이란?

- 정의
    - 데이터로부터 학습하도록 컴퓨터를 프로그래밍하는 과학(또는 예술)입니다.
    - 명시적인 프로그래밍 없이 컴퓨터가 학습하는 능력을 갖추게 하는 연구 분야
    - 어떤 작업 에 대한 컴퓨터 프로그램의 성능을 P로 측정했을 때 경험 E로 인해 성능이 향상됐다면, 이 컴퓨터 프로그램은 작업 T와 성능 측정 P에 대해 경험 E로 학습한 것
        - ex) T : 스팸메일 구분 / E : 훈련 데이터 / P : 측정 방법


### 1.2 왜 머신러닝을 사용하는가?

- 기존 솔루션으로는 많은 수동 조정과 규칙이 필요한 문제 : 하나의 머신러닝 모델이 코드를 간단하고 더 잘 수행되도록 할 수 있다.
- 전통적인 방식으로는 전혀 해결 방법이 없는 복잡한 문제 : 가장 뛰어난 머신러닝 기법으로 해결 방법을 찾을 수 있다.
- 유동적인 환경 : 머신러닝 시스템은 새로운 데이터에 적응할 수 있습니다.
- 복잡한 문제와 대량의 데이터에서 통찰 얻기


### 1.3 머신러닝 시스템의 종류

##### 1.3.1 지도학습과 비지도 학습

- 지도학습 : 알고리즘에 주입하는 훈련 데이터에 레이블이라는 원하는 답이 포함
    - 종류 : K-최근접 이웃, 선형 회귀, 로지스틱 회귀, SVM, 결정트리, 랜덤 포레스트, 신경망

- 비지도 학습 : 훈련 데이터에 레이블이 없는 것
    - 군집 : K-Means, 계츨 군집 분석, 기댓값 최대화
    - 시각화와 차원 축소 : PCA, 지역적 선형 임베딩, t-SNE
    - 연관 규칙 학습 : Apriori, Eclat

- 준지도 학습 : 레이블이 일부만 있는 데이터를 다루는 것
    - 심층 신뢰 신경망(DBN) : 여러 겹으로 쌓은 제한된 볼츠만 머신(RBM)에 기초한 신경망
    
- 강화 학습 : 학습하는 시스템을 에이전트라고 부르며 환경을 관찰해서 행동을 실행하고 그 결과로 보상 또는 벌점을 부여하는 방식.

##### 1.3.2 배치 학습과 온라인 학습

- 온라인 학습 : 연속적으로 데이터를 받고 빠른 변화에 스스로 적응해야 하는 시스템에 적합.
    - 미니배치 : 작은 묶음 단위로 주입해 시스템을 훈련 / 매 학습 단계가 빠르고 비용이 적게 든다.
    - 학습률 : 변화하는 데이터에 얼마나 빠르게 적응할 것인지.
    
##### 1.3.3 사례 기반 학습과 모델 기반 학습
    
- 사례 기반 학습 : 사례를 기억함으로써 학습 / 유사도 측정을 사용해 새로운 데이터에 일반화
- 모델 기반 학습 : 모델을 만들어 예측하는 것
    - 데이터 분석
    - 모델 선택
    - 훈련 데이터로 모델을 훈련
    - 새로운 데이터에 모델을 적용 및 예측
    
### 1.4 머신러닝의 주요 도전 과제

- 충분하지 않은 양의 훈련 데이터
- 대표성 없는 훈련 데이터
- 낮은 품질의 데이터
- 관련 없는 특성
    - 특성 선택 : 가지고 있는 특성 중에서 훈련에 가장 유용한 특성 선택
    - 특성 추출 : 특성을 결합해 더 유용한 특성을 만든다.
- 훈련 데이터 과대적합 / 과소적합

### 1.5 테스트와 검증

- 훈련 세트, 테스트 세트로 분류
- 검증세트, 교차 검증을 통해 과적합을 방지
